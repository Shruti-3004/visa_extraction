```python
from prettyprinter import pprint
import re
import json

with open('C:\\air_ticket\\arabic_page0.jpg_4_ara_eng_set_dpi-convert_to_grayscale.txt', 'r', encoding='utf-8') as fp:
    text = fp.read()
    text = text.splitlines()
    text = [line for line in text if not line.isspace() and len(line) > 0]

# pprint(text)

def get_authority(text):
    authority = ''
    for line in text:
        if 'authority' in line.lower():
            authority = line
            return authority

def get_date_of_issue(text):
    date = ''
    for line in text:
        if re.search('.*date.*of.*issue.*', line.lower()):
            words = line.split()
            ind = words.index(':')
            date = words[ind+1]
            return date

def get_place_of_issue(text):
    place = ''
    for line in text:
        if re.search('.*place.*of.*issue.*', line.lower()):
            words = line.split()
            ind = words.index(':')
            place = words[ind+2]
            return place

def get_validity(text):
    date = ''
    for line in text:
        if re.search('.*valid.*until.*', line.lower()):
            words = line.split()
            ind = words.index(':')
            date = words[ind+1]
            return date

def is_arabic(word):
    regex = r"[\u0600-\u06FF]+"
    if re.findall(regex, word):
        return True
    return False

def get_full_name(text):
    name = ''
    for line in text:
        if re.search('.*full.*name.*', line.lower()):
            words = line.split()
            ind = words.index(':')
            for word in words[ind+1:]:
                if is_arabic(word):
                    break
                name += word + ' '
            return name.strip()

def get_nationality(text):
    nationality = ''
    for line in text:
        if 'nationality' in line.lower():
            words = line.split()
            ind = words.index(':')
            nationality = words[ind+1]
            return nationality

def get_place_of_birth(text):
    place = ''
    flag = 0
    for line in text:
        if 'place of birth' in line.lower():
            words = line.split()
            for i in range(len(words)):
                if ':' in words[i]:
                    for j in range(i+1, len(words)):
                        if is_arabic(words[j]):
                            flag = 1
                            break
                        place += words[j] + ' '
                if flag:
                    return place

def get_date_of_birth(text):
    dob = ''
    for line in text:
        if 'date of birth' in line.lower():
            words = line.split()
            for i in range(len(words)):
                if ':' in words[i]:
                    dob = words[i+1]
                    return dob

def get_passport_no(text):
    pass_no = ''
    for line in text:
        if 'passport no.' in line.lower():
            line = line.replace('|', '')
            words = line.split()
            for i in range(len(words)):
                if ':' in words[i]:
                    pass_no = ' '.join(words[i+1:i+4])
                    return pass_no

def get_profession(text):
    profession = ''
    for line in text:
        if 'profession' in line.lower():
            words = line.split()
            for i in range(len(words)):
                if ':' in words[i]:
                    profession = words[i+1]
                    return profession

##---------------------------------------------##

def get_arabic_authority(text):
    for line in text:
        return line

def get_arabic_date_of_issue(text):
    doi = ''
    for line in text:
        # print(line)
        if re.search('.*date.*of.*issue.*', line.lower()):
            lline = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            # print(line)
            words = line.split()
            # print('words---> ', words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i+2]

def get_arabic_place_of_issue(text):
    for line in text:
        if re.search('.*place.*of.*issue.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i]

def get_arabic_validity_date(text):
    for line in text:
        if re.search('.*valid.*until.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i-1]

def get_arabic_full_name(text):
    for line in text:
        if re.search('.*full.*name.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return ' '.join(words[i+2:])

def get_arabic_nationality(text):
    for line in text:
        if re.search('.*nationality.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i+2]

def get_arabic_place_of_birth(text):
    for line in text:
        if re.search('.*place.*of.*birth.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return ' '.join(words[i+3:])

def get_arabic_date_of_birth(text):
    for line in text:
        if re.search('.*date.*of.*birth.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i-2]

def get_arabic_passport_no(text):
    for line in text:
        if re.search('.*passport.*no.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i-2]

def get_arabic_profession(text):
    for line in text:
        if re.search('.*profession.*', line.lower()):
            line = line.replace('\u200f', '')
            line = line.replace('\u200e', '')
            words = line.split()
            # print(words)
            for i in range(len(words)):
                if is_arabic(words[i]):
                    return words[i+2]

authority = get_authority(text)
date_of_issue = get_date_of_issue(text)
place_of_issue = get_place_of_issue(text)
validity_date = get_validity(text)
name = get_full_name(text)
nationality = get_nationality(text)
place_of_birth = get_place_of_birth(text)
date_of_birth = get_date_of_birth(text)
passport_no = get_passport_no(text)
profession = get_profession(text)

ar_authority = get_arabic_authority(text)
ar_date_of_issue = get_arabic_date_of_issue(text)
ar_place_of_issue = get_arabic_place_of_issue(text)
ar_validity_date = get_arabic_validity_date(text)
ar_name = get_arabic_full_name(text)
ar_nationality = get_arabic_nationality(text)
ar_place_of_birth = get_arabic_place_of_birth(text)
ar_date_of_birth = get_arabic_date_of_birth(text)
ar_passport_no = get_arabic_passport_no(text)
ar_profession = get_arabic_profession(text)
# print('Arabic date of issue---> ', ar_date_of_issue)


# print('Authority---> ', authority)
# print('Date of Issue---> ', date_of_issue)
# print('Place of Issue---> ', place_of_issue)
# print('Validity---> ', validity_date)
# print('Full Name---> ', name)
# print('Nationality---> ', nationality)
# print('Place of Birth---> ', place_of_birth)
# print('Date of Birth---> ', date_of_birth)
# print('Passport No---> ', passport_no)
# print('Profession---> ', profession)

result = {}

authority = authority.encode('ascii', 'ignore')
authority = authority.decode()


result['English'] = {
    'Authority': authority,
    'Date of Issue': date_of_issue,
    'Place of Issue': place_of_issue,
    'Validity': validity_date,
    'Full Name': name,
    'Nationality': nationality,
    'Place of Birth': place_of_birth,
    'Date of Birth': date_of_birth,
    'Passport Number': passport_no,
    'Profession': profession
}

f = open('keys.json', encoding='utf-8')
data = json.load(f)

result['Arabic'] = {
    data['Authority'][0]: ar_authority,
    data['Date of Issue'][0]: ar_date_of_issue,
    data['Place of Issue'][0]: ar_place_of_issue,
    data['Validity'][0]: ar_validity_date,
    data['Full Name'][0]: ar_name,
    data['Nationality'][0]: ar_nationality,
    data['Place of Birth'][0]: ar_place_of_birth,
    data['Date of Birth'][0]: ar_date_of_birth,
    data['Passport Number'][0]: ar_passport_no,
    data['Profession'][0]: ar_profession
}

pprint(result)
```
